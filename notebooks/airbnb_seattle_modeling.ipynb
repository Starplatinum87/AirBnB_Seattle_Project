# AirBnB Seattle Project - Modeling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.dummy import DummyRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.linear_model import LassoCV, RidgeCV, Ridge, Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
import time

import sys
sys.path.append("../scripts")

import airbnb_functions as abnb
```

```python
# Import data
listings = pickle.load(open('../data/pickles/listings_munged2.pkl', 'rb'))
```

# Prediction Target - Occupancy Rate
We're going to create a model that effectively predicts the occupancy rate of a property listing to gain understainding of the variables that most contribute to it. This can help direct AirBnB's efforts to educate hosts about about what contributes to successful properties, benefiting both hosts and AirBnB's bottom line. 

# Standard Scale Data
Since this is a regression problem, we'll standard scale the numerical data.

## Numerical columns to scale

```python
numerical_columns = ['host_listings_count',
 'host_total_listings_count',
 'accommodates',
 'bathrooms',
 'bedrooms',
 'beds',
 'price',
 'guests_included',
 'minimum_nights',
 'maximum_nights',
 'number_of_reviews',
 'review_scores_rating',
 'calculated_host_listings_count',
 'len_summary',
 'len_space',
 'len_description',
 'len_host_about',
 'host_response_rate_nanfilled']
```

```python
X, y = listings.drop('occupancy_rate', axis=1), listings.occupancy_rate
listings_scaled = abnb.scale_data(X, y, numerical_columns)
listings_scaled
```

# Create Training & Test Sets

```python
X, y = listings_scaled.drop('occupancy_rate', axis=1), listings.occupancy_rate
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# Create baseline prediction
Use the DummyRegressor from sklearn to create the simplest prediction, taking the mean of all occupancy_rates and making that the prediction.

## Dummy Regressor Prediction

```python
# Use the default 'mean' dummy regressor
base = DummyRegressor()
base.fit(X_train, y_train)
y_pred = base.predict(X_test)

print('Baseline Mean Squared Error:', round(mean_squared_error(y_test, y_pred), 4))
```

- We have our **baseline Mean Squared Error on test of 0.1178**

# Modeling

## Ridge Default
Use linear regression with Ridge regularization and GridSearchCV to find the best parameter.

```python
# We'll use the Ridge results to inform our parameter grid, as C is l2 regularization (Ridge)
params = {'alpha': [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0, 10000.0]}
ridge = Ridge()
ridgeGridSearch = GridSearchCV(ridge, param_grid=params, scoring='neg_mean_squared_error', n_jobs=-1, cv=10)
ridgeGridSearch.fit(X_train, y_train)
```

```python
ridgeGridSearch_results = pd.DataFrame(ridgeGridSearch.cv_results_)
```

```python
print('Mean Fit Time:', round(ridgeGridSearch.refit_time_, 4))
print('Best Parameters:', ridgeGridSearch.best_params_)
print('Best Mean Squared Error:', round((ridgeGridSearch.best_score_ * -1), 4))
```

```python
# MSE scores for C 
fig, ax = plt.subplots(figsize=[10,7])
ax.plot(np.log10(list(ridgeGridSearch_results['param_alpha'])), 
         ridgeGridSearch_results['mean_test_score'] * -1);
```

## Lasso Regression
Lasso may be more promising since it has the effect of performing feature selection through its regularization, and we have a lot of features compared to the amount of observations.

```python
params = {'alpha': [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0, 10000.0]}
lasso = Lasso()
lassoGridSearch = GridSearchCV(lasso, param_grid=params, scoring='neg_mean_squared_error', n_jobs=-1, cv=10)
lassoGridSearch.fit(X_train, y_train)
```

```python
lassoGridSearch_results = pd.DataFrame(lassoGridSearch.cv_results_)
```

```python
print('Mean Fit Time:', round(lassoGridSearch.refit_time_, 4))
print('Best Parameters:', lassoGridSearch.best_params_)
print('Best Mean Squared Error:', round((lassoGridSearch.best_score_ * -1), 4))
```

```python
fig, ax = plt.subplots(figsize=[10,7])
ax.plot(np.log10(list(lassoGridSearch_results['param_alpha'])), 
         lassoGridSearch_results['mean_test_score'] * -1);
```

### Ridge & Lasso OLS Conclusions
- **Ridge Best Model**
    - **Alpha: 1000**
    - **Mean Fit Time: 0.0205**
    - **Mean Squared Error: 0.1161**
- **Lasso Best Model**
    - **Alpha: 0.01**
    - **Mean Fit Time: 0.0222**
    - **Mean Squared Error: 0.1162**
- Ridge and Lasso ended up with roughly the same MSE, 0.1161 and 0.1162, as well as very similar fit times, 0.0222, and 0.0205.
- Interestingly though **neither outperformed the baseline mean occupancy rate by a significant degree**. **Baseline was 0.1178 and the Ridge and Lasso regressions were 0.1162.** 

## Support Vector Regression
There are many kernel options with Support Vector Machines. We'll give each a try with default values before tweaking hyperparameters on the most promising one. 

### SVR Linear Kernel

```python
start = time.time()
svrLinear = SVR(kernel='linear')
svrLinear_scores = cross_val_score(svrLinear, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
end = time.time()
```

```python
print('Fit Time: {}'.format(round(end-start, 4)))
print('Cross-Fold Score Array:', svrLinear_scores * -1)
print('Mean Squared Error:', np.mean(svrLinear_scores) * -1)
```

- Note that **model fitting took a much longer time than with Ridge and Lasso.**
- There are **two significant outlier scores affecting the mean MSE.**

```python
fig, ax = plt.subplots(figsize=[10,7])
sns.kdeplot((svrLinear_scores * -1), ax=ax)
```

### SVR Polynomial Kernel

```python
start = time.time()
svrPoly = SVR(kernel='poly')
svrPoly_scores = cross_val_score(svrPoly, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
end = time.time()
```

```python
print('Fit Time: {}'.format(round(end-start, 4)))
print('Cross-Fold Score Array:', svrPoly_scores * -1)
print('Mean Squared Error:', np.mean(svrPoly_scores) * -1)
```

```python
fig, ax = plt.subplots(figsize=[10,7])
sns.kdeplot((svrPoly_scores * -1), ax=ax)
```

- Faster fitting time than linear. Similar distribution to the linear kernel, with one major outlier score. 
- Worth noting that **the one major outlier score drastically affected the mean MSE.**

### SVR RBF Kernel

```python
start = time.time()
svrRBF = SVR(kernel='rbf')
svrRBF_scores = cross_val_score(svrRBF, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
end = time.time()
```

```python
print('Fit Time: {}'.format(round(end-start, 4)))
print('Cross-Fold Score Array:', svrRBF_scores * -1)
print('Mean Squared Error:', np.mean(svrRBF_scores) * -1)
```

```python
fig, ax = plt.subplots(figsize=[10,7])
sns.kdeplot((svrRBF_scores * -1), ax=ax)
```

- Similar fit time to polynomial, which was much faster that linear, and there are **no significant outlier folds from the cross-fold validation.** The **mean MSE is much lower than both, and each of the individual fold scores is generally lower than the previous two methods.** 

### SVR Sigmoid

```python
start = time.time()
svrSigmoid = SVR(kernel='sigmoid')
svrSigmoid_scores = cross_val_score(svrSigmoid, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
end = time.time()
```

```python
print('Fit Time: {}'.format(round(end-start, 4)))
print('Cross-Fold Score Array:', svrSigmoid_scores * -1)
print('Mean Squared Error:', np.mean(svrSigmoid_scores) * -1)
```

```python
fig, ax = plt.subplots(figsize=[10,7])
sns.kdeplot((svrSigmoid_scores * -1), ax=ax)
```

### SVR Default Values Conclusions
- The RBF kernel performed the best of the SVMs, but it's performance only matches that of Lasso and Ridge regularized OLS.
- However, we may be able to get better performance by tweaking the C and gamma hyperparameters. We'll use GridSeachCV to try this.

```python
# MSE scores for C with gamma locked at 0.1
fig, ax = plt.subplots(figsize=[10,7])
ax.plot(np.log10(list(rbfGridSearch_results['param_C'].loc[rbfGridSearch_results['param_gamma'] == 0.1])), 
         rbfGridSearch_results['mean_test_score'].loc[rbfGridSearch_results['param_gamma'] == 0.1] * -1);
```

## Random Forest Regression
Random Forests are an ensemble modeling method that takes several weak learners and combines their results to make predictions. We'll use GridSearchCV again to try to find the best parameters. 

Random Forests have many more hyperparameters than SVM or OLS methods, but the two that are the most important are the number of decision trees (`n_estimators`) and the number of features (`max_features`). We'll focus on these in our searches. 

### `n_estimators` Search

```python
# We'll start with n_estimators 
params = {'n_estimators': [100, 200, 500, 1000, 2000, 5000]}
rfr = RandomForestRegressor()
rfrGridSearch1 = GridSearchCV(rfr, param_grid=params, scoring='neg_mean_squared_error', n_jobs=-1, cv=10)
rfrGridSearch1.fit(X_train, y_train)
```

```python
rfrGridSearch1_results = pd.DataFrame(rfrGridSearch1.cv_results_)
```

```python
print('Mean Fit Time:', round(rfrGridSearch1.refit_time_, 4))
print('Best Parameters:', rfrGridSearch1.best_params_)
print('Best Mean Squared Error:', round((rfrGridSearch1.best_score_ * -1), 4))
```

```python
# MSE scores for n_estimators
fig, ax = plt.subplots(figsize=[10,7])
ax.plot(list(rfrGridSearch1_results['param_n_estimators']), 
         rfrGridSearch1_results['mean_test_score'] * -1);
```
